```python
import pandas as pd
d1 = {"No of Elements":[5000,10000,15000,20000,25000],"Time taken":[0.02,0.03,0.05,0.08,0.09]}
d2 = pd.DataFrame(d1)
d2
```

```python
# @title No of Elements vs Time taken

from matplotlib import pyplot as plt
d2.plot(kind='scatter', x='No of Elements', y='Time taken', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
